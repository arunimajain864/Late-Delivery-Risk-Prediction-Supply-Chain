### Create Risk Dataset

```python
risk_df = X_test.copy()

risk_df['Actual_Risk'] = y_test.values

risk_df['Risk_Probability'] = risk_scores
```

### Risk Category

```python
def classify_risk(prob):

    if prob < 0.30:
        return "Low Risk"

    elif prob < 0.70:
        return "Medium Risk"

    else:
        return "High Risk"
```

```python
risk_df['Risk_Category'] = risk_df['Risk_Probability'].apply(classify_risk)
```
